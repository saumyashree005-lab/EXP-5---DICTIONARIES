#**Creation of Dictionaries**

```python
thisdict = {
    "brand" : "Ford",
    "model" : "Mustang",
    "year" : 1964,
}
print(thisdict)
```

#**Retriving Values From Dictionary**

```python
thisdict = {
    "brand" : "Ford",
    "model" : "Mustang",
    "year" : 1964,
}
print(thisdict["brand"])
```

#**Overwriting Data In A Dictionary**

```python
#dictionary takes latest value given for key. it will not allow duplicates
thisdict = {
    "brand" : "Ford",
    "model" : "Mustang",
    "year" : 1964,
    "year" : 2020,
}
print(thisdict)
```

#**Changing/Adding Key:value and Class Of Dictionary**

```python
car = {
    "brand" : "Ford",
    "model" : "Mustang",
    "year" : 1964,
}
car ["colour"] = "red"
print(car)

x = car.values()
print(x)

print(type(car))
```

#**Removing data from Dictionary**

```python
student = {
    "roll" : 101,
    "name" : "soham",
    "branch" : "Entc",
}
student.pop("branch")
print("After Removal : ",student)
```
